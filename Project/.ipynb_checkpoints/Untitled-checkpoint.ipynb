# Step 1: Data Cleaning & Preprocessing

```python
# Data Features
sensor_data.info()
```

```python
# Data Understanding
sensor_data.describe()
```

```python
# Data-Preprocessing

#removing column 'sensor_15' because it has all null value.
sensor_data = sensor_data.drop('sensor_15', 1)
sensor_data = sensor_data.drop('Unnamed: 0', 1)

sensor_data.shape
```

```python
sensor_data['date'] = sensor_data['timestamp'].apply(lambda x: x.split(' ')[0])
sensor_data['time'] = sensor_data['timestamp'].apply(lambda x: x.split(' ')[1])
sensor_data = sensor_data.drop(['timestamp'], 1)
```

```python
sensordata_ipd = sensor_data.fillna(sensor_data.median())
```

```python
sensordata_ipd.shape
```

```python
sensordata_ipd.hist(figsize=(20,20))
```

```python
# Removing Outliers Using Zscore
z_scores = zscore(sensordata_ipd.iloc[:,:51])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
sdata_olr = sensordata_ipd[filtered_entries]
```

```python
x1,y1 = sensor_data.shape
x2,y2 = sdata_olr.shape

removed_data_percent = ((x1-x2)/x1)*100

print("In total "+ str(round(removed_data_percent,2)) +"% of data were removed after outliers eradication")
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i, j]) >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset

    return dataset
```

# Step 3: Model Development

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X = correlation(sdata_std,0.7)
```

```python
le = LabelEncoder()
Y = le.fit_transform(sdata_olr['machine_status'])
len(np.zero(Y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42)
```

```python
h = .02  # step size in the mesh

# Calculate min, max and limits
x_min, x_max = X.iloc[:, 0 ].min() - 1, X.iloc[:, 0 ].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc [:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Put the result into a color plot
plt.figure()
plt.scatter(X.iloc[:, 0], X.iloc[:, 1])
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("Data points")
plt.show()
```

Linear Regression - Model 

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
regr_pred=regr.predict(X_test)
```

```python
acc_regr=round(regr.score(X_train,y_train)*100,10)
acc_regr
```

```python
mse = r2_score(y_test, regr_pred)
mse
```

```python
# sb.set_style('whitegrid')
# sb.lmplot(x ='Actual', y ='Predicted', data = df_lr)
# sb.regplot(y_test,regr_pred)

plt.figure(figsize=(5,5))
plt.scatter(y_test, regr_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(regr_pred), max(y_test))
p2 = min(min(regr_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

Logistic Regression

```python
#applying logistic regression
lr = LogisticRegression(C = 0.2)
clf1 = lr.fit(X_train, y_train)
y_pred1 = clf1.predict(X_test)
```

```python
acc_log_reg=roc_auc_score(y_test, y_pred1)*100
acc_log_reg
```
